```python
import os
import pandas as pd

# 
# 0. SET WORKING DIRECTORY
# 
os.chdir(r"D:\Data Sets")

print("Files:", os.listdir())

# 
# Q1. Gold buying vs USD Index
# 

# ---- Gold data ----
gold = pd.read_excel("Gold_UNComtrade_2019_2024.xlsx")

gold_year = (
    gold
    .groupby("Year")["TradeValue_USD"]
    .sum()
    .reset_index()
)

# ---- USD Index data ----
usd = pd.read_csv("USD_Index_DXY.csv")

# Auto-detect date column
date_col = [c for c in usd.columns if 'date' in c.lower()][0]

usd[date_col] = pd.to_datetime(usd[date_col])
usd["Year"] = usd[date_col].dt.year

usd_year = (
    usd
    .groupby("Year")["DTWEXBGS"]
    .mean()
    .reset_index()
)

# ---- Merge & Correlation ----
gold_usd = pd.merge(gold_year, usd_year, on="Year")

print("\nQ1 RESULT: Gold vs USD Index Correlation")
print(gold_usd[["TradeValue_USD", "DTWEXBGS"]].corr())
```

```python
import os
import pandas as pd

# ----
# SET WORKING DIRECTORY
# ----
os.chdir(r"D:\Data Sets")

# ==========================================
# Q2. Does higher crypto settlement correlate
#     with smaller cargo sizes?
# ==========================================

# ---- Load Bitcoin trading volume ----
btc = pd.read_excel("Trading_Volume.xlsx")

# Detect date column
btc_date_col = [c for c in btc.columns if "date" in c.lower() or "time" in c.lower()][0]

# Detect BTC volume column (numeric, not date)
btc_volume_col = [
    c for c in btc.columns
    if c != btc_date_col and pd.api.types.is_numeric_dtype(btc[c])
][0]

btc[btc_date_col] = pd.to_datetime(btc[btc_date_col])
btc["Year"] = btc[btc_date_col].dt.year

btc_year = (
    btc
    .groupby("Year")[btc_volume_col]
    .sum()
    .reset_index()
    .rename(columns={btc_volume_col: "BTC_Volume"})
)

# ---- Load Crude Oil cargo data ----
oil = pd.read_excel("Crude_OIL-Data-2019_2024.xlsx")

oil_year = (
    oil
    .groupby("Year")["netWgt"]
    .sum()
    .reset_index()
    .rename(columns={"netWgt": "Cargo_Weight"})
)

# ---- Merge & Correlation ----
crypto_oil = pd.merge(btc_year, oil_year, on="Year")

print("\nQ2 RESULT: Crypto Settlement vs Cargo Size Correlation")
print(crypto_oil[["BTC_Volume", "Cargo_Weight"]].corr())
```

```python
import os
import pandas as pd

# ------------------------------------------
# SET WORKING DIRECTORY
# ------------------------------------------
os.chdir(r"D:\Data Sets")

# ==========================================
# Q3. BRICS Currency Share – Pre vs Post UAE Policy
# ==========================================

swift = pd.read_csv("SWIFT_RMB_Tracker.csv")

# Auto-detect BRICS/RMB share column (numeric, not Year)
share_col = [
    c for c in swift.columns
    if c.lower() != "year" and pd.api.types.is_numeric_dtype(swift[c])
][0]

print("Using currency share column:", share_col)

# Create policy period
swift["Policy_Period"] = swift["Year"].apply(
    lambda x: "Pre-2023" if x < 2023 else "Post-2023"
)

# Pre vs Post comparison
brics_share = (
    swift
    .groupby("Policy_Period")[share_col]
    .mean()
    .reset_index()
)

print("\nQ3 RESULT: BRICS Currency Share (Pre vs Post UAE Policy)")
print(brics_share)
```
